```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_phase_full(u, L, d, freq, x):
    Ltot = np.sqrt(L**2 + u**2) + np.sqrt(d**2 + (u-x)**2)
    offset = 2*np.pi*freq*(L + np.sqrt(d**2+x**2))/c # phi(0)
    t = (1/c) * (const-((u**2)/(2*f)))
    return 2*np.pi*freq* (Ltot/c + t) - offset


def get_phase_Taylor(u,L,d,freq,x):
    t = (1/c) * (const-((u**2)/(2*f)))
    offset = L+d
    return L*(1+(u**2)/(2*L**2)) + d*(1+(u**2)/(2*(d**2))) + c*t - offset
```

```python

f = .25
lam = 500e-09
L = 1

c=3e8
const = 0
freq = c/lam
```

# Part a

```python
u = np.linspace(-.065,.065,100)
d1=1.0/3
d2=.4
phase1 = get_phase_Taylor(u,L,d1,freq,0)
phase2 = get_phase_Taylor(u,L,d2,freq,0)


plt.figure()
plt.plot(u, phase1, 'b')
plt.plot(u,phase2, 'g--')
plt.xlabel('position within aperture, u (m)')
plt.ylabel('relative phase')
plt.legend(["Focused", "Unfocused"])
plt.title("Lens-Based Imaging Focusing using Taylor Approximation")
plt.show()
```
