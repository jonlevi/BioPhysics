```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
%matplotlib inline
```

```python

B=1
nw = 1.34

y_vals = np.linspace(.75,1,15)
theta_2 = np.arcsin(y_vals/B)
theta_1 = theta_2
theta_3 = theta_2


psi = np.arcsin(np.sin(theta_2)/nw)
alpha = np.pi-2*psi
gamma = (2*theta_2+2*alpha)-np.pi

```

```python
def pol2cart(rho, phi):
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    return(x, y)

def pol2cartX(rho, phi):
    x = rho * np.cos(phi)
    return x

def pol2cartY(rho, phi):
    y = rho * np.sin(phi)
    return y
```

```python
Ax = -2*np.ones(len(y_vals))
Ay = y_vals

Px = pol2cartX(B,np.pi-theta_2)
Py = pol2cartY(B,np.pi-theta_2)



Qx = pol2cartX(B,np.pi-theta_2-alpha)
Qy = pol2cartY(B,np.pi-theta_2-alpha)



Rx = pol2cartX(B,np.pi-theta_2-2*alpha)
Ry = pol2cartY(B,np.pi-theta_2-2*alpha)




fig, ax = plt.subplots(figsize=(15, 15))
axisx = np.linspace(-2,2,10)
axisy = np.zeros(10)
plt.plot(axisx,axisy, 'k--')
plt.scatter(Px,Py, c='k')
plt.scatter(Qx,Qy, c='k')
plt.scatter(Rx,Ry, c='k')
plt.scatter(Ax,Ay, c='k')
# plt.scatter(Sx, Sy, c='k')
for i in range(len(Px)):
#     plt.plot(Sx,Sy,'o')
    plt.plot([Ax[i], Px[i]], [Ay[i], Py[i]], 'k')
    plt.plot([Px[i],Qx[i]], [Py[i], Qy[i]], 'k')
    plt.plot([Qx[i],Rx[i]], [Qy[i], Ry[i]], 'k')

slopes = np.tan(np.pi-gamma)    
for j in range(len(Rx)):
    Sx = np.linspace(-2, Rx[j],3)
    Sy = slopes[j]*(Sx-Rx[j])+Ry[j]
    plt.plot(Sx,Sy,'k')
    
circle = Circle((0, 0), B, facecolor='none',
                edgecolor=(0, 0, 0), linewidth=3, alpha=0.5)
ax.add_patch(circle)

plt.xlim([-2.2,2])
plt.ylim([-2.2,2])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ray-Optics for Spherical Water Droplet')
plt.show()
```

```python
plt.figure()
plt.plot(y_vals,gamma, 'k')
plt.xlabel('$Y_0$ (Radians)')
plt.ylabel('$\gamma$ (a.u.)')
plt.title('$\gamma$ as a function of vertical displacement')
plt.show()
```
