# Part b

```python
p = 1e-6
wavelen = 532e-9
d = 24.5e-2
psi_list = np.zeros(10000, dtype='complex')
x_vals = np.linspace(-.5,.5,10000)

# sum loop over all terms in sum (2-40)
    # get all the terms in the psi equation
    # x loop (sum over linspace)
for i,x in enumerate(x_vals):
    total = 0
    
    for n in range(2,22,2):
        L_plus = np.sqrt((d**2+(x-((n-1)*p)/2)**2))
        L_minus = np.sqrt((d**2+(x+((n-1)*p)/2)**2))
        psi = np.e**(1j * 2*np.pi * L_plus/wavelen) + np.e**(1j * 2*np.pi * L_minus/wavelen)
        
        total = total + psi

    psi_list[i] = (abs(total))**2

```

```python
plt.figure()
plt.plot(x_vals, psi_list)
plt.xlabel("Position on Screen (m)")
plt.ylabel("Probabiltity Amplitude Squared")
plt.show()
```

## Comments
Increasing the size of p turns this problem back into a regular slit diffraction problem, and we begin to see the regular diffraction pattern emerge with peaks going further out. Instead of just seeing the 2 side peaks around the 0 peak, we now see more and more peaks as p gets larger and larger
